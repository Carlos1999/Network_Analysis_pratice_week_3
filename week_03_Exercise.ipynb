## Question 12

Consider a bipartite network of N nodes, $N_1$ nodes of type 1 and $N_2$ nodes of type 2 (so that $N_1 + N_2 = N$). What is the maximum number of links in this network?

**Answer:** Temos que:
$$
  N_1 + N_2 = N \implies  
  N_2 = N - N_1
$$  
Além disso, sabemos também que:
$$
    L_{max} = N_1 \cdot N_2 = N_1(N - N_1) = -N_{1}^2 + N \cdot N_{1} \\
$$
Perceba que $-N_{1}^2 + N \cdot N_{1}$ representa uma função do segundo grau, cujo gráfico é uma parábola. Como o coeficiente de $N_{1}^2$ é menor que $0$, nesse caso $-1$, sabemos que a parábola dessa função quadrática possui a concavidade voltada para baixo, isto é, admite um ponto máximo. Logo, para saber qual a quantidade máxima de links que tal rede bipartida admite precisamos calcular a cordenada $y$ do ponto máximo dessa parábola.

Para isso, utilizaremos que o $x$ do vértice $x_v$ = $\frac{-b}{2a}$. Analisando a equação, vemos que $b = N$ e $a = -1$. Substituindo temos: 
$$
  x_v = \frac{-b}{2a} = \frac{-N}{2(-1)} = \frac{-N}{-2} = \frac{N}{2}\\
$$
Por fim, para descobrirmos o $y$ do vértice $y_v$, isto é, a quantidade máxima de links que a rede pode admitir, basta substituir o $N_1$ da equação por $x_v = \frac{N}{2}$. Assim,
$$
 y_v = -(\frac{N}{2})^2 + N(\frac{N}{2}) = \frac{-N^2}{4} + \frac{N^2}{2} = \frac{N^2}{4}\\
$$
Logo, a quantidade máxima de liniks que esse tipo de rede pode ter é $\frac{N^2}{4}$.

## Hands On 01


Often in the context of trees, a node with degree 1 is called a *leaf*. Write a function named `get_leaves` that takes a graph as an argument, loops through the nodes, and returns a list of nodes with degree 1.

```python
import networkx as nx
%matplotlib inline

G = nx.Graph()
G.add_edges_from([
        ('a', 'b'),
        ('a', 'd'),
        ('c', 'd'),
    ])


```

```python
def get_leaves(G):
  # cria uma lista para armazenar os nós folha
  lista_folhas = []
  # percorre nós e armazena na lista os de grau 1
  for i in G.nodes:
    if(G.degree(i) == 1):
      lista_folhas.append(i)
  # retorna lista de folhas
  return lista_folhas
```

```python
# this block of code is only for networkx adapt it for graph-tool
G = nx.Graph()
G.add_edges_from([
        ('a', 'b'),
        ('a', 'd'),
        ('c', 'd'),
    ])
assert set(get_leaves(G)) == {'c', 'b'}
```

## Hands On 02


Write a function max_degree that takes a graph as its argument, and returns a 2-tuple with the name and degree of the node with highest degree.

```python
def max_degree(G):
    # define grau do primeiro nó como máximo
    max = G.degree(list(G.nodes)[0])
    # cria lista que armazena nó de maior grau, nome e grau
    node_max_degree = [0, 0]

    for node in G.nodes:
      # troca o valor da lista se o grau do nó for maior que o máximo até o momento
      if(G.degree(node)>=max):
        max = G.degree(node)
        node_max_degree[0] = node
        node_max_degree[1] = G.degree(node)
    
    # converte para tupla e retorna
    return tuple(node_max_degree)
```

## Hands on 03


Write a function `mutual_friends` that takes a graph and two nodes as arguments, and returns a list (or set) of nodes that are linked to both given nodes. For example, in the graph `SG` drawn above,

    mutual_friends(SG, 'Alice', 'Claire') == ['Frank']

an empty list or set should be returned in the case where two nodes have no mutual friends, e.g. George and Bob in `SG` drawn above.

```python

def mutual_friends(G, node_1, node_2):
  # cria lista para armazenas os vizinhos mútuos de node_1 e node_2
  lista_mutual_friends = []
  
  for node in G.nodes:
    # pega vizinhos de cada nó
    vizinhos = list(G.neighbors(node))
    # adiciona na lista se for vizinho mútuo
    if(node_1 in vizinhos and node_2 in vizinhos):
      lista_mutual_friends.append(node)
  
  return lista_mutual_friends
```

```python
# this block of code is only for networkx adapt it for graph-tool
SG = nx.read_adjlist('friends.adjlist')
assert mutual_friends(SG, 'Alice', 'Claire') == ['Frank']
assert mutual_friends(SG, 'George', 'Bob') == []
assert sorted(mutual_friends(SG, 'Claire', 'George')) == ['Dennis', 'Frank']
```
